```python
import pandas as pd
import numpy as np
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_squared_error
from math import sqrt

# Cargar los datos de la pestaña "Demanda_Historica"
demanda_historica = pd.read_excel('TRABAJO-FINAL- Pronósticos.xlsx', sheet_name='Demanda_Historica')

# Filtrar los SKUs que son AX, BX y CX (asumiendo que ya tienes la clasificación ABC y XYZ)
# Aquí debes reemplazar `merged_data` con el DataFrame que ya tiene las clasificaciones ABC y XYZ
AX_BX_CX_SKUs = merged_data[(merged_data['Categoria_ABC'] == 'A') & (merged_data['Categoria_XYZ'] == 'X') |
                            (merged_data['Categoria_ABC'] == 'B') & (merged_data['Categoria_XYZ'] == 'X') |
                            (merged_data['Categoria_ABC'] == 'C') & (merged_data['Categoria_XYZ'] == 'X')]['SKU']

# Función para realizar el pronóstico con suavizamiento exponencial triple
def forecast_holt_winters(series, periods=18):
    model = ExponentialSmoothing(series, trend='add', seasonal='add', seasonal_periods=12)
    fit = model.fit(optimized=True, remove_bias=True)
    forecast = fit.forecast(periods)
    return fit, forecast

# Función para realizar el pronóstico con promedios móviles
def forecast_moving_average(series, window=12, periods=18):
    moving_avg = series.rolling(window=window).mean().dropna()
    last_value = moving_avg.iloc[-1]
    forecast = pd.Series([last_value] * periods, index=range(len(series), len(series) + periods))
    return forecast

# Función para calcular el RMSE
def calculate_rmse(actual, predicted):
    return sqrt(mean_squared_error(actual, predicted))

# Iterar sobre cada SKU AX, BX y CX
for sku in AX_BX_CX_SKUs:
    # Obtener la serie de tiempo del SKU
    sku_series = demanda_historica[demanda_historica['SKU'] == sku].iloc[:, 1:].values.flatten()

    # Pronóstico con Holt-Winters
    hw_fit, hw_forecast = forecast_holt_winters(sku_series)

    # Pronóstico con promedios móviles
    ma_forecast = forecast_moving_average(pd.Series(sku_series))

    # Comparar los modelos con RMSE (usando los últimos 12 meses para validación)
    train, test = sku_series[:-12], sku_series[-12:]

    # Ajustar el modelo Holt-Winters con los datos de entrenamiento
    hw_fit_train = ExponentialSmoothing(train, trend='add', seasonal='add', seasonal_periods=12).fit(optimized=True, remove_bias=True)
    hw_pred = hw_fit_train.forecast(12)

    # Calcular RMSE para Holt-Winters
    rmse_hw = calculate_rmse(test, hw_pred)

    # Calcular RMSE para promedios móviles
    ma_pred = forecast_moving_average(pd.Series(train), window=12, periods=12)
    rmse_ma = calculate_rmse(test, ma_pred)

    # Seleccionar el mejor modelo
    if rmse_hw < rmse_ma:
        best_forecast = hw_forecast
        best_model = "Holt-Winters"
    else:
        best_forecast = ma_forecast
        best_model = "Promedios Móviles"

    # Mostrar el pronóstico del mejor modelo
    print(f"SKU: {sku}")
    print(f"Mejor modelo: {best_model} (RMSE: {min(rmse_hw, rmse_ma):.2f})")
    print("Pronóstico a 18 meses:")
    print(best_forecast)
    print("\n")
    ("El análisis ABC, XYZ, ADI y CV² ha sido completado y los resultados se han guardado en 'Clasificacion_ABC_XYZ_ADI_CV2_SKUs.xlsx'.")
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_squared_error
from math import sqrt

# Cargar los datos de las pestañas "Demanda_Historica" y "Precios_Costos"
demanda_historica = pd.read_excel('TRABAJO-FINAL- Pronósticos.xlsx', sheet_name='Demanda_Historica')
precios_costos = pd.read_excel('TRABAJO-FINAL- Pronósticos.xlsx', sheet_name='Precios_Costos')

# Calcular la demanda total anual para cada SKU
demanda_historica['Demanda_Total_Anual'] = demanda_historica.iloc[:, 1:].sum(axis=1)

# Obtener solo las columnas de demanda mensual para cálculos de desviación estándar
demanda_mensual_columns = demanda_historica.iloc[:, 1:-1] # Excluir 'SKU' y 'Demanda_Total_Anual'

# Calcular la desviación estándar de la demanda mensual para cada SKU
demanda_historica['Desviacion_Estandar'] = demanda_mensual_columns.std(axis=1)

# Calcular la demanda promedio mensual para cada SKU
num_months = demanda_mensual_columns.shape[1]
demanda_historica['Demanda_Promedio_Mensual'] = demanda_historica['Demanda_Total_Anual'] / num_months

# Fusionar los datos de demanda histórica (con nuevas columnas) y precios/costos
merged_data = pd.merge(demanda_historica[['SKU', 'Demanda_Total_Anual', 'Desviacion_Estandar', 'Demanda_Promedio_Mensual']],
                       precios_costos, on='SKU')

# Calcular el valor de consumo anual (Demanda_Total_Anual * Costo_Unitario)
merged_data['Valor_Consumo_Anual'] = merged_data['Demanda_Total_Anual'] * merged_data['Costo_Unitario']

# Ordenar los SKUs por valor de consumo anual de mayor a menor
merged_data = merged_data.sort_values(by='Valor_Consumo_Anual', ascending=False)

# Calcular el valor acumulado y el porcentaje acumulado
merged_data['Valor_Acumulado'] = merged_data['Valor_Consumo_Anual'].cumsum()
merged_data['Porcentaje_Acumulado'] = 100 * merged_data['Valor_Acumulado'] / merged_data['Valor_Consumo_Anual'].sum()

# Clasificar los SKUs en categorías A, B y C
merged_data['Categoria_ABC'] = pd.cut(
    merged_data['Porcentaje_Acumulado'],
    bins=[0, 60, 80, 100],
    labels=['A', 'B', 'C'],
    include_lowest=True
)

# Calcular el coeficiente de variación (Desviación Estándar / Demanda Promedio Mensual)
merged_data['Coeficiente_Variacion'] = merged_data['Desviacion_Estandar'] / merged_data['Demanda_Promedio_Mensual']

# Clasificar los SKUs en categorías X, Y, Z según el coeficiente de variación
merged_data['Categoria_XYZ'] = pd.cut(
    merged_data['Coeficiente_Variacion'],
    bins=[0, 0.25, 1.0, float('inf')],
    labels=['X', 'Y', 'Z'],
    include_lowest=True
)

# Calcular el ADI (Intervalo Promedio de Demanda)
merged_data['ADI'] = num_months / (demanda_mensual_columns > 0).sum(axis=1).reindex(merged_data.index) # Reindex to align

# Calcular el CV² (Coeficiente de Variación al Cuadrado)
merged_data['CV2'] = merged_data['Coeficiente_Variacion'] ** 2

# Filtrar los SKUs que son AX, BX y CX
AX_BX_CX_SKUs = merged_data[
    ((merged_data['Categoria_ABC'] == 'A') & (merged_data['Categoria_XYZ'] == 'X')) |
    ((merged_data['Categoria_ABC'] == 'B') & (merged_data['Categoria_XYZ'] == 'X')) |
    ((merged_data['Categoria_ABC'] == 'C') & (merged_data['Categoria_XYZ'] == 'X'))
]['SKU']

# Función para realizar el pronóstico con suavizamiento exponencial triple
def forecast_holt_winters(series, periods=18):
    model = ExponentialSmoothing(series, trend='add', seasonal='add', seasonal_periods=12)
    fit = model.fit(optimized=True, remove_bias=True)
    forecast = fit.forecast(periods)
    return fit, forecast

# Función para realizar el pronóstico con promedios móviles
def forecast_moving_average(series, window=12, periods=18):
    moving_avg = series.rolling(window=window).mean().dropna()
    last_value = moving_avg.iloc[-1]
    forecast = pd.Series([last_value] * periods, index=range(len(series), len(series) + periods))
    return forecast

# Función para calcular el RMSE
def calculate_rmse(actual, predicted):
    return sqrt(mean_squared_error(actual, predicted))

# Lista para almacenar los resultados de los pronósticos
forecast_results = []

# Iterar sobre cada SKU AX, BX y CX
for sku in AX_BX_CX_SKUs:
    # Obtener la serie de tiempo del SKU
    sku_series = demanda_historica[demanda_historica['SKU'] == sku].iloc[:, 1:1+num_months].values.flatten() # Get only monthly data

    # Pronóstico con Holt-Winters
    hw_fit, hw_forecast = forecast_holt_winters(pd.Series(sku_series))

    # Pronóstico con promedios móviles
    ma_forecast = forecast_moving_average(pd.Series(sku_series))

    # Comparar los modelos con RMSE (usando los últimos 12 meses para validación)
    train, test = sku_series[:-12], sku_series[-12:]

    # Ajustar el modelo Holt-Winters con los datos de entrenamiento
    hw_fit_train = ExponentialSmoothing(train, trend='add', seasonal='add', seasonal_periods=12).fit(optimized=True, remove_bias=True)
    hw_pred = hw_fit_train.forecast(12)

    # Calcular RMSE para Holt-Winters
    rmse_hw = calculate_rmse(test, hw_pred)

    # Calcular RMSE para promedios móviles
    ma_pred = forecast_moving_average(pd.Series(train), window=12, periods=12)
    rmse_ma = calculate_rmse(test, ma_pred)

    # Seleccionar el mejor modelo
    if rmse_hw < rmse_ma:
        best_forecast = hw_forecast
        best_model = "Holt-Winters"
        best_rmse = rmse_hw
    else:
        best_forecast = ma_forecast
        best_model = "Promedios Móviles"
        best_rmse = rmse_ma

    # Guardar los resultados
    forecast_results.append({
        'SKU': sku,
        'Mejor Modelo': best_model,
        'RMSE': best_rmse,
        'Pronóstico 18 Meses': best_forecast.tolist()
    })

# Crear un DataFrame con los resultados de los pronósticos
forecast_df = pd.DataFrame(forecast_results)

# Guardar los resultados en un archivo Excel
with pd.ExcelWriter('Pronosticos_AX_BX_CX.xlsx') as writer:
    forecast_df.to_excel(writer, sheet_name='Pronósticos', index=False)

print("El análisis de pronósticos ha sido completado y los resultados se han guardado en 'Pronosticos_AX_BX_CX.xlsx'.")
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
from math import sqrt

# Cargar los datos de las pestañas "Demanda_Historica" y "Precios_Costos"
demanda_historica = pd.read_excel('TRABAJO-FINAL- Pronósticos.xlsx', sheet_name='Demanda_Historica')
precios_costos = pd.read_excel('TRABAJO-FINAL- Pronósticos.xlsx', sheet_name='Precios_Costos')

# Calcular la demanda total anual para cada SKU
demanda_historica['Demanda_Total_Anual'] = demanda_historica.iloc[:, 1:].sum(axis=1)

# Fusionar los datos de demanda y precios/costos
merged_data = pd.merge(demanda_historica[['SKU', 'Demanda_Total_Anual']], precios_costos, on='SKU')

# Calcular el valor de consumo anual (Demanda_Total_Anual * Costo_Unitario)
merged_data['Valor_Consumo_Anual'] = merged_data['Demanda_Total_Anual'] * merged_data['Costo_Unitario']

# Ordenar los SKUs por valor de consumo anual de mayor a menor
merged_data = merged_data.sort_values(by='Valor_Consumo_Anual', ascending=False)

# Calcular el valor acumulado y el porcentaje acumulado
merged_data['Valor_Acumulado'] = merged_data['Valor_Consumo_Anual'].cumsum()
merged_data['Porcentaje_Acumulado'] = 100 * merged_data['Valor_Acumulado'] / merged_data['Valor_Consumo_Anual'].sum()

# Clasificar los SKUs en categorías A, B y C
merged_data['Categoria_ABC'] = pd.cut(
    merged_data['Porcentaje_Acumulado'],
    bins=[0, 60, 80, 100],
    labels=['A', 'B', 'C'],
    include_lowest=True
)

# Calcular la desviación estándar de la demanda mensual para cada SKU
demanda_mensual = demanda_historica.iloc[:, 1:-1]  # Excluir la columna 'Demanda_Total_Anual'
demanda_mensual['Desviacion_Estandar'] = demanda_mensual.std(axis=1)

# Fusionar la desviación estándar con los datos anteriores
merged_data = pd.merge(merged_data, demanda_mensual[['Desviacion_Estandar']], left_index=True, right_index=True)

# Calcular el coeficiente de variación (Desviación Estándar / Demanda Promedio Mensual)
merged_data['Demanda_Promedio_Mensual'] = merged_data['Demanda_Total_Anual'] / 36  # 36 meses de datos
merged_data['Coeficiente_Variacion'] = merged_data['Desviacion_Estandar'] / merged_data['Demanda_Promedio_Mensual']

# Clasificar los SKUs en categorías X, Y, Z según el coeficiente de variación
merged_data['Categoria_XYZ'] = pd.cut(
    merged_data['Coeficiente_Variacion'],
    bins=[0, 0.25, 1.0, float('inf')],
    labels=['X', 'Y', 'Z'],
    include_lowest=True
)

# Calcular el ADI (Intervalo Promedio de Demanda)
merged_data['ADI'] = 36 / (demanda_mensual > 0).sum(axis=1)

# Calcular el CV² (Coeficiente de Variación al Cuadrado)
merged_data['CV2'] = merged_data['Coeficiente_Variacion'] ** 2

# Filtrar los SKUs que son AY, BY y CY
AY_BY_CY_SKUs = merged_data[
    ((merged_data['Categoria_ABC'] == 'A') & (merged_data['Categoria_XYZ'] == 'Y')) |
    ((merged_data['Categoria_ABC'] == 'B') & (merged_data['Categoria_XYZ'] == 'Y')) |
    ((merged_data['Categoria_ABC'] == 'C') & (merged_data['Categoria_XYZ'] == 'Y'))
]['SKU']

# Función para realizar el pronóstico con promedio móvil
def forecast_moving_average(series, window=12, periods=18):
    moving_avg = series.rolling(window=window).mean().dropna()
    last_value = moving_avg.iloc[-1]
    forecast = pd.Series([last_value] * periods, index=range(len(series), len(series) + periods))
    return forecast

# Función para realizar el pronóstico con el método de Croston
def forecast_croston(series, periods=18):
    # Implementación simplificada del método de Croston
    non_zero_demand = series[series > 0]
    if len(non_zero_demand) == 0:
        return pd.Series([0] * periods, index=range(len(series), len(series) + periods))

    avg_demand = non_zero_demand.mean()
    avg_interval = len(series) / len(non_zero_demand)

    forecast = pd.Series([avg_demand / avg_interval] * periods, index=range(len(series), len(series) + periods))
    return forecast

# Función para calcular el RMSE
def calculate_rmse(actual, predicted):
    return sqrt(mean_squared_error(actual, predicted))

# Lista para almacenar los resultados de los pronósticos
forecast_results = []

# Iterar sobre cada SKU AY, BY y CY
for sku in AY_BY_CY_SKUs:
    # Obtener la serie de tiempo del SKU
    sku_series = demanda_historica[demanda_historica['SKU'] == sku].iloc[:, 1:].values.flatten()

    # Pronóstico con promedio móvil
    ma_forecast = forecast_moving_average(pd.Series(sku_series))

    # Pronóstico con Croston
    croston_forecast = forecast_croston(pd.Series(sku_series))

    # Comparar los modelos con RMSE (usando los últimos 12 meses para validación)
    train, test = sku_series[:-12], sku_series[-12:]

    # Pronóstico con promedio móvil para los datos de entrenamiento
    ma_pred = forecast_moving_average(pd.Series(train), window=12, periods=12)

    # Pronóstico con Croston para los datos de entrenamiento
    croston_pred = forecast_croston(pd.Series(train), periods=12)

    # Calcular RMSE para promedio móvil
    rmse_ma = calculate_rmse(test, ma_pred)

    # Calcular RMSE para Croston
    rmse_croston = calculate_rmse(test, croston_pred)

    # Seleccionar el mejor modelo
    if rmse_ma < rmse_croston:
        best_forecast = ma_forecast
        best_model = "Promedio Móvil"
        best_rmse = rmse_ma
    else:
        best_forecast = croston_forecast
        best_model = "Croston"
        best_rmse = rmse_croston

    # Guardar los resultados
    forecast_results.append({
        'SKU': sku,
        'Mejor Modelo': best_model,
        'RMSE': best_rmse,
        'Pronóstico 18 Meses': best_forecast.tolist()
    })

# Crear un DataFrame con los resultados de los pronósticos
forecast_df = pd.DataFrame(forecast_results)

# Guardar los resultados en un archivo Excel
with pd.ExcelWriter('Pronosticos_AY_BY_CY.xlsx') as writer:
    forecast_df.to_excel(writer, sheet_name='Pronósticos', index=False)

print("El análisis de pronósticos ha sido completado y los resultados se han guardado en 'Pronosticos_AY_BY_CY.xlsx'.")
```
